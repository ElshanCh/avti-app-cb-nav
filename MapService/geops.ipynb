```python
import pyproj

def convert_coordinates(x, y, from_crs='EPSG:3857', to_crs='EPSG:4326'):
    # Define the input and output coordinate reference systems (CRS)
    crs_from = pyproj.CRS(from_crs)
    crs_to = pyproj.CRS(to_crs)

    # Create a transformer object to perform the conversion
    transformer = pyproj.Transformer.from_crs(crs_from, crs_to, always_xy=True)

    # Perform the coordinate transformation
    lon, lat = transformer.transform(x, y)

    # Return the converted coordinates as a tuple
    return  lat,lon

bb = [45.127055490565446, 7.714534030655944,45.126979435153345, 7.7150101377565266]
bb[0], bb[1] = convert_coordinates(bb[0], bb[1],from_crs='EPSG:4326', to_crs='EPSG:3857')
bb[2], bb[3] = convert_coordinates(bb[2], bb[3],from_crs='EPSG:4326', to_crs='EPSG:3857')
bb = list(map(int, bb))
bb
z ="10"
BBOX = f"BBOX {' '.join(str(value) for value in bb)} {z}"
BBOX
# # Define the coordinates you want to convert
# x = 853442
# y = 5626367

# # Call the function to convert the coordinates
# converted_lon, converted_lat = convert_coordinates(x, y)

# # Output the converted coordinates
# print(converted_lon, ",", converted_lat)
```

```python
import pyproj

def convert_coordinate_column(column):
    # Define the input and output coordinate reference systems (CRS)
    epsg3857 = pyproj.CRS('EPSG:3857')
    epsg4326 = pyproj.CRS('EPSG:4326')

    # Create a transformer object to perform the conversion
    transformer = pyproj.Transformer.from_crs(epsg3857, epsg4326, always_xy=True)

    # Convert each coordinate pair in the column
    converted_column = []
    for coordinate_pair in column:
        converted_coordinates = []
        for coordinate in coordinate_pair:
            x, y = coordinate
            lon, lat = transformer.transform(x, y)
            converted_coordinates.append([lon, lat])
        converted_column.append(converted_coordinates)

    return converted_column
```

```python
import pandas as pd
import requests
import itertools
import pandas as pd
import json

# Read the app.log file
with open('logs/app_20230510170548.log', 'r') as file:
    log_lines = file.readlines()

# List to store valid JSON lines
valid_lines = []

# Iterate over the log lines
for line in log_lines:
    try:
        # Parse JSON
        data = json.loads(line)

        # Check if the line structure is similar to the second example
        if data["source"] == "trajectory":
            valid_lines.append(data)
    except json.JSONDecodeError:
        # Skip lines that are not valid JSON
        continue

# Normalize JSON lines until train_id value
df = pd.json_normalize(valid_lines)

df_x = df[[ 'content.geometry.type', 'content.geometry.coordinates','content.properties.bounds','content.properties.gen_range', 'content.properties.tenant','content.properties.type', 'content.properties.time_intervals','content.properties.train_id', 'content.properties.event_timestamp','content.properties.line.id', 'content.properties.line.name','content.properties.line.color', 'content.properties.line.text_color','content.properties.line.stroke', 'content.properties.timestamp','content.properties.state','content.properties.route_identifier','content.properties.delay']]

MY_ACCESS_TOKEN = "5cc87b12d7c5370001c1d6550e7752961af14d3fb1b9b3cff77e26f9"
base_url = "https://api.geops.io/tracker-http/v1/journeys/"

# Create an empty DataFrame with column names
columns = ["properties.publisher","properties.publisherUrl","properties.operator","properties.operatorUrl","properties.train_id","geometry.geometries"]
df_to_fill = pd.DataFrame(columns=columns)

def create_nested_dict(row, column_name, value):
    if '.' in column_name:
        main_node, subnodes = column_name.split('.', 1)
        if main_node not in row:
            row[main_node] = {}
        create_nested_dict(row[main_node], subnodes, value)
    else:
        row[column_name] = value


for train_id  in df_x['content.properties.train_id']:
    # Construct the URL for the GET request
    url = f"{base_url}{train_id}/"
    # print(url)
    # Set the headers for the request
    headers = {
        'accept': 'application/json',
        'Authorization': MY_ACCESS_TOKEN
    }

    # Make the GET request
    response = requests.get(url, headers=headers)

    # Process the response as needed
    if response.status_code == 200:
        # Successful request
        data = response.json()
        # print(data)
        # Process the data further
        # Normalize JSON lines until train_id value
        df_full = pd.json_normalize(data)
        df_features = pd.json_normalize(df_full.explode(["features"])["features"])
        merged_df = pd.merge(df_full[["properties.publisher","properties.publisherUrl","properties.operator","properties.operatorUrl","properties.train_id"]],
                            df_features[["geometry.geometries","properties.train_id"]],
                            on="properties.train_id")
        df_to_fill = pd.concat([df_to_fill,merged_df], ignore_index=True)

    else:
        # Error occurred during the request
        print(f"Error: {response.status_code} - {response.text}")

df_x = pd.merge(df_x, df_to_fill, left_on='content.properties.train_id', right_on='properties.train_id')
df_x.drop('properties.train_id', axis=1, inplace=True)


# Convert DataFrame to list of JSON objects
json_list = []
for _, row in df_x.iterrows():
    json_obj = {}
    for column_name, value in row.items():
        create_nested_dict(json_obj, column_name, value)
    json_list.append(json_obj)

# Print the list of JSON objects
for json_obj in json_list:
    print(json.dumps(json_obj))
```
